```python
def create_gradient_image(width: int, height: int) -> np.ndarray:
    """
    Create a gradient image with three channels (RGB).

    Args:
        width (int): Width of the image.
        height (int): Height of the image.

    Returns:
        np.ndarray: Gradient image with three channels (RGB).
    """
    x = np.linspace(0, 255, width, dtype=np.uint8)
    y = np.linspace(0, 255, height, dtype=np.uint8)
    xv, yv = np.meshgrid(x, y)

    gradient_image = np.zeros((height, width, 3), dtype=np.uint8)
    gradient_image[..., 0] = xv  # Red channel gradient
    gradient_image[..., 1] = yv  # Green channel gradient
    gradient_image[..., 2] = (xv + yv) // 2  # Blue channel gradient

    return gradient_image

# Sample usage
image = create_gradient_image(100, 100)  # Gradient pattern image
center = (70, 10)  # Center coordinates
crop_size = 40  # Crop size
cropped_image = crop_image(image, center, crop_size)
print(cropped_image.shape)
fig, ax = plt.subplots(1, 2)
ax[0].imshow(image)
ax[1].imshow(cropped_image)
```
